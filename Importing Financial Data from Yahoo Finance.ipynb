```python
import yfinance as yf

stocks = ['AAPL', 'MSFT', 'GOOGL']
data = yf.download(stocks, start="2025-01-01", end="2025-03-31", auto_adjust=False)

# Accessing the 'Adj Close' data
print(data['Adj Close'].head())
print(data['Adj Close'].tail())
```

```python
data['Adj Close'].plot(figsize=(12, 8), title='Adjusted Close Prices of Stocks')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.show()
```
