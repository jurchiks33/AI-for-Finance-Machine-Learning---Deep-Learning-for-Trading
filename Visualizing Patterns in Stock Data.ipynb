# Trend Identification

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch online stock data for Microsoft from Yahoo Finance
msft_data = yf.download('MSFT', start='2020-01-01', end='2024-03-27')

# Simple trend visualization
msft_data['Close'].plot(figsize=(12, 6), title='Microsoft Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.show()
```

# Head and Shoulders Pattern

```python
# Fetch stock data for MSFT for given dates
data = yf.download('MSFT', start='2021-07-01', end='2022-08-01')

# Plotting the specific area 
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price', linewidth=2)

# Head & Shoulders 
points = [
    ('2021-09-07', 'Left Shoulder'),
    ('2021-12-27', 'Head'),
    ('2022-04-04', 'Right Shoulder')
]

for date_str, label in points:
    date = pd.to_datetime(date_str)
    price = data.loc[date, 'Close'].item()
    plt.annotate(label,
                 xy=(date, price),
                 xytext=(date, price + 12),
                 arrowprops=dict(facecolor='red', arrowstyle='->'),
                 fontsize=12, color='green')

# Visualization
plt.title('Head and Shoulders Pattern for MSFT (2021-2022)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

# Double Tops and Bottoms

```python
# Fetch stock data for a Double Top pattern in MSFT
data = yf.download('MSFT', start='2021-10-01', end='2022-02-01')

# Plot Closing Prices
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price', linewidth=2)

# Annotate the Double Tops
double_top_dates = ['2021-11-19', '2021-12-27']
for date_str in double_top_dates:
    date = pd.to_datetime(date_str)
    price = data.loc[date, 'Close'].item()
    plt.annotate('Double Top',
                 xy=(date, price),
                 xytext=(date, price + 8),
                 arrowprops=dict(facecolor='blue', arrowstyle='->'),
                 fontsize=12, color='darkred')

# Visualization
plt.title('Microsoft Double Top Pattern (Late 2021)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

# Triangle Patterns

```python
import matplotlib.patches as patches
import matplotlib.dates as mdates

# Fetch MSFT stock data
data = yf.download('MSFT', start='2021-03-01', end='2021-09-01')

# Plot closing prices
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price', linewidth=2)

# Adjusted trendlines to drawing
upper_dates = ['2021-04-27', '2021-06-09']
upper_prices = [254, 240]

lower_dates = ['2021-05-12', '2021-06-09']
lower_prices = [230, 240]

# Plotting trendlines
plt.plot(pd.to_datetime(upper_dates), upper_prices, 'r-', linewidth=2, label='Upper Trendline')
plt.plot(pd.to_datetime(lower_dates), lower_prices, 'r-', linewidth=2, label='Lower Trendline')

# Convert date to Matplotlib float format for Polygon
triangle = patches.Polygon([
    (mdates.date2num(pd.to_datetime('2021-05-12')), 230),  # bottom-left
    (mdates.date2num(pd.to_datetime('2021-04-28')), 254),  # bottom-right
    (mdates.date2num(pd.to_datetime('2021-06-09')), 240)   # tip
], closed=True, color='grey', alpha=0.3)

plt.gca().add_patch(triangle)

# Visualization 
plt.title('Microsoft Triangle Pattern', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

# Wedge Patterns

```python
import matplotlib.dates as mdates

# Visualization of wedge pattern
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Closing Price')

# wedge trendline points matching real price action
wedge_dates = pd.to_datetime(['2021-03-15', '2021-08-01'])
upper_wedge = [243, 284]   # Upper trendline wedge
lower_wedge = [210, 260]   # Lower trendline wedge

# Drawing wedge pattern trendlines
plt.plot(wedge_dates, upper_wedge, 'g--', linewidth=2, label='Upper Wedge Line')
plt.plot(wedge_dates, lower_wedge, 'r--', linewidth=2, label='Lower Wedge Line')

# Filling the wedge
plt.fill_between(wedge_dates, lower_wedge, upper_wedge, color='yellow', alpha=0.3)

# Visualisation 
plt.title('Microsoft Wedge Pattern', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
